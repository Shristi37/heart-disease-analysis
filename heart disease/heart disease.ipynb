```python
data['target'].value_counts()
```

```python
plt.bar(data['target'].index, data['target'])

# Set x-axis label
plt.xlabel('Index')

# Set y-axis label
plt.ylabel('Target Variable')

# Set plot title
plt.title('Bar Plot of Target Variable')

# Convert the Matplotlib plot to a Seaborn countplot
sns.despine()
sns.countplot(x=data['target'], color='skyblue')

# Show the plot
plt.show()
```

```python
data['sex'].value_counts()
```

```python
sns.countplot(x=data['sex'])

# Set x-axis label
plt.xlabel('Sex')

# Set y-axis label
plt.ylabel('Count')

# Set plot title
plt.title('Count Plot of Sex')

# Show the plot
plt.show()
```

```python
sns.countplot(x='sex',hue="target",data=data)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

```python
sns.distplot(data['age'],bins=20)
plt.show()
```

```python
sns.countplot(x=data['cp'])

# Customize x-axis ticks and labels
plt.xticks([0, 1, 2, 3], ["Typical Angina", "Atypical Angina", "Non-anginal Pain", "Asymptomatic"])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Set x-axis label
plt.xlabel('Chest Pain Type')

# Set y-axis label
plt.ylabel('Count')

# Set plot title
plt.title('Count Plot of Chest Pain Type')

# Show the plot
plt.show()
```

```python
sns.countplot(x="cp",hue="target",data=data)
plt.legend(labels=["no-disease","disease"])
plt.show()
```

```python
sns.countplot(x="fbs",hue="target",data=data)
plt.legend(labels=["no-disease","disease"])
plt.show()
```
